Q2: We plan to use the average of the absolute differences in the common movies watched between every pair as our distance metric. The pair with the lowest average difference of all the common movies watched would then be the movie soul mate of the given user.

Here we chose to not consider the number of common movies watched because we believe having watched a few movies but with similar ratings is a better indicator than having watched many movies but with a large variance. Also to consider the variance we switched to the average rather than just simply comparing the sum of the absolute differences

Pseudo code:

1. Fetch all the details from the movies file and store in movies list
2. Fetch all the details from the ratings file and store in ratings list
3. For every movie in movies list
    Go throuh every rating for that movie in the ratings list
        Add this rating to the total
        increment the count of ratings
    Append this total ratings divided by count of ratings to a new average ratings list. This will act as global average.
4. Input the name of the user and store this as username
5. Check if this username is valid or not by comparing in the first column of the ratings list, if not exit
6. go through the usernames row in ratings list and append the movie positions with 0 into a new list not_watched
7. For every pair of usernames in ratingslist
        Compare every rating from not_watched that has non-zero for the other person
            sum the differences and divide by the count of such movies to create average
            append this average to a new list movie_averages
8. Sort this list by the soulmates rating
9. Return the top 5 results if the list is long, otherwise return the list in the required format.

 

```python
from typing_extensions import final
import csv 
# import copy

# pseudo-code:
# find Q2's anwser at 3-5.

# 1. Get the averages of all the movies and the count of how many ratings each movie has got.
# 2. input a username, for that user get the list of movies this person has not watched.

# 3. calculate the difference of the user's rating with a certain other user's, only if they both have watched that movie.
# 4. get the 'soul mate score' by finding the absolute sum of the differences and dividing it by the movie number watched by both users.
# 5. find the soul mate by finding the lowest 'soulmate score'

# 6. find the movies that the user has not watched but the soul mate has, sort them according to the soul mate's ratings
# 7. print the soul mate's name, the top 5 movies that the user has not watched but the soul mate has, the average scores, the ratings

newList = []
total = 0

# Fetch all movie names from the file
def getMovieNames():
	
	with open('movies.csv', newline='', encoding='utf-8') as f: 
		reader = csv.reader(f)

		movie_names = []

		for row in reader: 
			movie_names.append(row[0])
	f.close()

	movie_names = movie_names[1:]
	return movie_names

# Fetch ratings of people from ratings.csv
def getRatingsList():
	ratingsList = []
	# Logic
	with open('ratings.csv', newline='', encoding='utf-8') as f: 
		reader = csv.reader(f)
	
		ratings = []

		for row in reader:
			ratings.append(row)
	f.close()
	return ratings
 
def getMovieAverages(movie_names, ratings_list):
	movie_averages = []
	counter =0

	# For each movie
	for movie_number in range(len(movie_names)):
		# print(ratings_list[0][counter+1])
		
		total = 0
		count_of_ratings = 0

		# Go through all the ratings for this movie
		for row in ratings_list:
			if int(row[movie_number+1]) != 0:
				total += int(row[movie_number + 1])
				count_of_ratings+=1
		
		movie_averages.append([movie_names[movie_number], total/count_of_ratings, count_of_ratings])
		
	return movie_averages

def getMoviesNotWatched(name_of_person):
	movie_numbers_not_watched = []
	for row in ratings_list:
		if row[0]!= name_of_person:
			continue
		
		for i in range(1, len(ratings_list[0])):
			if int(row[i])==0:
				movie_numbers_not_watched.append(i-1)
		break

	return movie_numbers_not_watched

def calculateScore(rating_one, rating_two):
	score = 0
	count = 0

	# Just taking the difference
	for i in range(1,len(rating_one)):
		if int(rating_one[i])==0 or int(rating_two[i])==0:
			continue
		count += 1
		score += abs(int(rating_one[i]) - int(rating_two[i]))
	
	return score/count

def getSoulMateInfo(movie_names,ratings_list,name_of_person):
	soul_mate_name=""
	soul_mate_score= float('inf')
	soul_mate_record=[]

	soul_mate_score_list =[]
	rating_of_person =[]

	# Find the score of the person
	for i in range(len(ratings_list)):
		if ratings_list[i][0] == name_of_person:
			rating_of_person = ratings_list[i]
			break

	# Compare this score to all the other scores
	for i in range(len(ratings_list)):
		if ratings_list[i][0] == name_of_person:
			continue
		else:
			current_soul_mate_score = calculateScore(ratings_list[i],rating_of_person)
			if current_soul_mate_score < soul_mate_score:
				soul_mate_score = current_soul_mate_score
				soul_mate_name = ratings_list[i][0]
				soul_mate_record = ratings_list[i]
	
	return soul_mate_name,soul_mate_score, rating_of_person, soul_mate_record

def getSoulMateUnwatechedMovies(person_record, soul_mate_record, movie_averages):
  final_list=[]
  not_watched_list = []
  for i in range(len(person_record)):
    if i==0:
      continue
    if int(person_record[i])==0:
      not_watched_list.append(i)
  for i in not_watched_list:
    if int(soul_mate_record[i])!=0:
      ma = movie_averages[i-1]
      ma.append(soul_mate_record[i])
      final_list.append(ma)
  return final_list
   
	

# Main Logic

movie_names = getMovieNames()
ratings_list = getRatingsList()
movie_averages = getMovieAverages(movie_names, ratings_list)
name_of_person = input("Please enter the name of the person: ")

#exit if the name is not there
isThere = False
for i in range(len(ratings_list)):
	if name_of_person == (ratings_list[i][0]):
		isThere = True
		break
if isThere==False:
	print("A new user!")
	print("Recommendations for", name_of_person, "using Algorithm 2:")
	movie_averages = sorted(movie_averages, key= lambda x: x[1],reverse=True)
	for row in movie_averages[:5]:
		print(row[0],', ', row[1], "(",row[2], " ratings)")
	exit()

else:

    movie_numbers_not_watched = getMoviesNotWatched(name_of_person)


    newList =[]

    for i in movie_numbers_not_watched:
            newList.append(movie_averages[i])


    name_of_soul_mate, soul_mate_distance, person_record, soul_mate_record = getSoulMateInfo(movie_names,ratings_list,name_of_person)

    list_and_rating_of_unwatched_soul_mate_movies = getSoulMateUnwatechedMovies(person_record, soul_mate_record, movie_averages)

    list_and_rating_of_unwatched_soul_mate_movies = sorted(list_and_rating_of_unwatched_soul_mate_movies, key=lambda x: x[3], reverse=True)



    print("Number of movies not yet watched by", name_of_person , ':', len(newList))
    print(name_of_person,"'s", "movie soul-mate is:", name_of_soul_mate , "(distance:", soul_mate_distance, ")")

    print("Recommendations for",name_of_person, "using Algorithm 2:") 

    for record in list_and_rating_of_unwatched_soul_mate_movies:
      print(record[0]," ",record[1], "(" ,record[2], " ratings);", "soul-mate’s rating:", record[3])

```
